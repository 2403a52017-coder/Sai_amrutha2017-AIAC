```python
import requests

def get_movie_details(title):
    api_key = "8ed5e5e4"  # â… IMPORTANT: Replace this with your actual OMDb API key
    url = f"http://www.omdbapi.com/?t={title}&apikey={api_key.strip()}"
    try:
        response = requests.get(url, timeout=5)
        response.raise_for_status()
        data = response.json()

        if data.get("Response") == "False":
            print("âŒ Movie not found. Please check the title.")
            return

        print("\nâœ± Movie Details:")
        print(f"Title       : {data.get('Title')}")
        print(f"Year        : {data.get('Year')}")
        print(f"Genre       : {data.get('Genre')}")
        print(f"IMDb Rating : {data.get('imdbRating')}")
        print(f"Director    : {data.get('Director')}")

    except requests.exceptions.Timeout:
        print("âš ï¸ Request timed out. Try again later.")
    except requests.exceptions.HTTPError:
        print("âš ï¸ HTTP error occurred. Check your API key or connection.")
    except requests.exceptions.RequestException:
        print("âš ï¸ Network error. Please check your internet connection.")

# â—Š Ask user for input
movie_title = input("Enter a movie title: ")
get_movie_details(movie_title)
```

```python
import json
from tabulate import tabulate

def fetch_arrival_times(station_id):
    """Simulate fetching live arrival data from a public transport API."""
    # Mock data (simulating an API response)
    mock_data = {
        "101": [
            {"route": "B12", "destination": "Central Station", "arrival_time": "2 mins"},
            {"route": "B15", "destination": "City Mall", "arrival_time": "5 mins"},
            {"route": "B22", "destination": "Airport", "arrival_time": "8 mins"},
            {"route": "B07", "destination": "Tech Park", "arrival_time": "12 mins"},
            {"route": "B09", "destination": "University", "arrival_time": "15 mins"},
            {"route": "B33", "destination": "Old Town", "arrival_time": "20 mins"}
        ],
        "202": [
            {"route": "T05", "destination": "North Station", "arrival_time": "3 mins"},
            {"route": "T07", "destination": "South Station", "arrival_time": "7 mins"},
            {"route": "T12", "destination": "East Junction", "arrival_time": "10 mins"},
            {"route": "T09", "destination": "Central Park", "arrival_time": "14 mins"},
            {"route": "T18", "destination": "Downtown", "arrival_time": "18 mins"}
        ]
    }

    # Error handling
    if station_id not in mock_data:
        raise ValueError("Invalid station ID or data not available.")

    # Return only the next 5 arrivals
    return mock_data[station_id][:5]

def display_arrivals(arrivals):
    """Display arrival times in a neat table."""
    table = [[item["route"], item["destination"], item["arrival_time"]] for item in arrivals]
    headers = ["Route Number", "Destination", "Arrival Time"]
    print(tabulate(table, headers, tablefmt="fancy_grid"))

# ---- Main Program ----
try:
    station_id = input("Enter Station ID (e.g., 101 or 202): ")
    arrivals = fetch_arrival_times(station_id)
    print(f"\nNext 5 arrivals for station {station_id}:\n")
    display_arrivals(arrivals)

except ValueError as e:
    print(f"Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
import yfinance as yf
from tabulate import tabulate

def get_stock_data(ticker):
    """Fetch daily stock price data for a given ticker symbol."""
    try:
        # Download data for the latest day
        stock = yf.Ticker(ticker)
        data = stock.history(period="1d")

        # Handle invalid ticker or empty response
        if data.empty:
            print("âŒ Invalid ticker symbol or no data available.")
            return

        # Extract required fields
        open_price = data['Open'].iloc[0]
        close_price = data['Close'].iloc[0]
        high = data['High'].iloc[0]
        low = data['Low'].iloc[0]
        volume = data['Volume'].iloc[0]

        # Display in a neat table
        table = [
            ["Open Price", f"${open_price:.2f}"],
            ["Close Price", f"${close_price:.2f}"],
            ["High", f"${high:.2f}"],
            ["Low", f"${low:.2f}"],
            ["Volume", int(volume)]
        ]

        print(f"\nğŸ“ˆ Stock Data for {ticker.upper()}:\n")
        print(tabulate(table, headers=["Field", "Value"], tablefmt="fancy_grid"))

    except Exception as e:
        print(f"âš ï¸ An error occurred: {e}")

# ---- Main Program ----
ticker_symbol = input("Enter the stock ticker symbol (e.g., AAPL, MSFT, GOOGL): ")
get_stock_data(ticker_symbol)
```

```python
import requests
import time

def translate_text(text, target_lang, retries=2):
    if not text.strip():
        print("âš  Error: Input text cannot be empty.")
        return

    # Two fallback translation APIs
    apis = [
        "https://libretranslate.com/translate",  # more reliable
        "https://api.mymemory.translated.net/get"  # fallback
    ]

    for attempt in range(retries):
        for api_url in apis:
            try:
                if "libretranslate" in api_url:
                    payload = {
                        "q": text,
                        "source": "auto",
                        "target": target_lang,
                        "format": "text"
                    }
                    headers = {"Content-Type": "application/json"}
                    response = requests.post(api_url, json=payload, headers=headers, timeout=10)
                    response.raise_for_status()
                    data = response.json()
                    translated = data.get("translatedText")

                else:  # MyMemory API fallback
                    response = requests.get(
                        api_url,
                        params={"q": text, "langpair": f"en|{target_lang}"},
                        timeout=10
                    )
                    response.raise_for_status()
                    data = response.json()
                    translated = data.get("responseData", {}).get("translatedText")

                if not translated:
                    raise ValueError("Empty translation or invalid response")

                print("\nğŸŒ Translation Result")
                print("===============================")
                print(f"Original Text : {text}")
                print(f"Translated ({target_lang}) : {translated}")
                print("===============================")
                return

            except Exception as e:
                print(f"âš  API failed ({api_url.split('//')[1].split('/')[0]}): {e}")
                time.sleep(1)

        print(f"â° Retry attempt {attempt + 1}/{retries}...\n")

    print("âŒ All translation attempts failed. Please try again later.")

# === Run the script ===
text = input("Enter text to translate: ")
target_lang = input("Enter target language code (e.g., 'es' for Spanish, 'fr' for French, 'hi' for Hindi): ")
translate_text(text, target_lang)
```
